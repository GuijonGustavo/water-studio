```python
cols = ['SAAM','OD_mg/L','COLI_TOT','pH_CAMPO','TEMP_AGUA','NI_TOT','E_COLI','HG_TOT','COLI_FEC','PB_TOT']

#nombre_sitio = df_Metztitlan["NOMBRE DEL SITIO"].array[1]

df_rows_eti = df_eti[df_eti['CLAVE PARÁMETRO'] == cols[1]]
titulo = df_rows_eti['NOMBRE DEL PARÁMETRO']
titulo = 'Nombre del parámetro: '+df_rows_eti.iloc[0,1]
#subtitulo = 'Nombre del sitio: '+nombre_sitio

df_res['SAAM'] = pd.to_numeric(df_res['SAAM'], downcast="float",errors='coerce')

df_clave_sitio = df_sit[df_sit["CUERPO DE AGUA"].str.contains("METZTITLAN")]
key_array_df = df_clave_sitio["CLAVE SITIO"]
for i in range (key_array_df.shape[0]):
#for i in range (1):
    nombre_sitio = df_Metztitlan["NOMBRE DEL SITIO"].array[i]
    subtitulo = 'Nombre del sitio: '+nombre_sitio
    
    key_array_df = df_cuerpo_de_agua["CLAVE SITIO"].array[i]
    serie = df_res[df_res["CLAVE SITIO"].str.contains(key_array_df)]
    longitud_tiempo = serie.shape[0]
    
    serie['SAAM'].replace('\< | \>', '')
    serie['SAAM'].dropna()
    
    rcParams['figure.figsize'] = 18, 8
    plt.figure()
    plt.title(titulo, fontsize=24)
    plt.suptitle(subtitulo, fontsize=24)
    plt.plot(serie['FECHA'], serie['SAAM'])
    plt.scatter(serie['FECHA'], serie['SAAM'], marker='o',color='red', s=40)
    plt.xticks(np.arange(0, longitud_tiempo, step=longitud_tiempo/6), fontsize=14) #Numberlines, nnumberlines/6
    plt.yticks(fontsize=14)
    plt.ylabel(df_rows.iloc[0,2], fontsize=14)
    plt.grid()
    plt.rcParams.update({'figure.max_open_warning': 200})   
    
```
